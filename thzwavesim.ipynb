```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
c = 299792458
lam = 800*10**-9
omega = 2*np.pi*c/(lam*10**12)
```

```python
omega
```

```python
t = np.linspace(0,0.6,1000)
```

```python
f = np.exp(-2*np.log(2)*((t-0.3)/0.1)**2)*np.cos(omega*t)
```

```python
plt.figure()
plt.plot(t,f, 'r')
plt.xlabel('Time(ps)')
plt.ylabel('Electric Field a.u.')
plt.savefig('nearirwave.svg')
plt.show()
```
